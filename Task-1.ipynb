```python
import pandas as pd

file_path = r"C:\Users\vishw\Downloads\archive (7)\Unemployment in India.csv"
df = pd.read_csv(file_path)

print("Dataset Preview:")
print(df.head())

print("Column Names:", df.columns)
df.columns = df.columns.str.strip().str.lower()

df.rename(columns={"date": "Date", "estimated unemployment rate (%)": "Unemployment Rate"}, inplace=True)

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

df.dropna(subset=['Date'], inplace=True)
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(x=df['Date'], y=df['Unemployment Rate'], hue=df['Region'], marker='o')
plt.title("Unemployment Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.grid()
plt.show()
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['Area'], y=df['Unemployment Rate'])
plt.title("Comparison of Unemployment Rate (Urban vs. Rural)")
plt.xlabel("Area Type")
plt.ylabel("Unemployment Rate (%)")
plt.show()
numerical_cols = ['Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate (%)']

plt.figure(figsize=(10, 6))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Employment Factors")
plt.show()
from statsmodels.tsa.arima.model import ARIMA

df_state = df[df['Region'] == "Maharashtra"].sort_values(by="Date")

model = ARIMA(df_state['Unemployment Rate'], order=(1,1,1))
model_fit = model.fit()

df_state['Predicted'] = model_fit.predict()
```
